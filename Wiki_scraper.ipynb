```python
import pandas as pd
import numpy as np
```

```python
from bs4 import BeautifulSoup
import re
import requests
```

```python
URL = 'https://en.wikipedia.org/wiki/Category:Indian_politicians_by_city_or_town'

page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
l = []
for a in soup.find_all('a', {'href':re.compile('^/wiki/Category')}):
    l.append(a['href'])
l[:4]
```

```python
l1 = []
for i in l:
    page = requests.get('https://en.wikipedia.org/{}'.format(i), timeout=10)
    soup = BeautifulSoup(page.content, 'html.parser')
    
    for a in soup.find_all('a', {'href':re.compile('^/wiki/Category')}):
        l1.append(a['href'])
```

```python
l2 = []
for i in l1:
    page = requests.get('https://en.wikipedia.org/{}'.format(i), timeout=10)
    soup = BeautifulSoup(page.content, 'html.parser')
    
    for a in soup.find_all('a', {'href':re.compile('^/wiki/')}):
        with open('links_final.txt','a') as f:
            f.write(a['href']+'\n')
```

```python
list_of_info = []
```

```python
def extractinfo(page):
    page = requests.get('https://en.wikipedia.org/{}'.format(page), timeout=10)
    # page = requests.get(articleUrl_final, timeout=10)
    soup = BeautifulSoup(page.content, 'html.parser')

    try:
        try:
            global row_img
            for img in soup('td',{'class':'infobox-image'}):
                img_link = img.find_all('a',{'class':'image'})
                for i in img_link:
                    row_img = i.find('img')['src']
        except Excepation as e:
            row_img=e

        try:
            born = soup.find_all('span',{'style':'display:none'})[0].text
        except Exception as e:
            born = 0

        try:
            died = soup.find_all('span',{'style':'display:none'})[1].text
        except Exception as e:
            died = 0

        dict_final = {
            'Name' : soup.find('h1',{'class':'firstHeading mw-first-heading'}).text,
            'born' : born,
            'died': died,
            'img_link' : 'https:{}'.format(row_img)
        }
        list_of_info.append(dict_final)

    except Exception as e:
            print(e)
```

```python
list_links = []
with open('links_final.txt','r') as file:
    wiki = file.read()
    list_links.append(wiki.split('\n'))
list_final = np.concatenate(list_links)
```

```python
for i in list_final:
    print(f"Running for article {i}")
    extractinfo(i)
```

```python
data = pd.DataFrame(list_of_info)
data.to_excel('output.xlsx')
```
